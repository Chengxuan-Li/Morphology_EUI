```python
import momepy
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
buildings = gpd.read_file('BuildingsWithEnergy.geojson')
```

```python
buildings
```

```python
plt.scatter(
    buildings['geometry'].area,
    buildings['TotalEnergyConsumptionCurrent_sum']/buildings['TotalFloorArea_sum'])
```

```python
import numpy as np
import pandas as pd
```

```python
d = {'AREA': buildings['geometry'].area, 'ENER': buildings['TotalEnergyConsumptionCurrent_sum']/buildings['TotalFloorArea_sum']}
df = pd.DataFrame(data=d)
df.dropna()
```

```python
normalized_df=(df-df.mean())/df.std()
```

```python
normalized_df.cov() # covariance
```

```python
normalized_df.corr() # correlation coefficient
```
