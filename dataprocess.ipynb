```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    'certificates.csv'
)
df
```

```python
df.dropna(axis = 0, subset = ['UPRN'])
```

```python
df = df[(df.ENERGY_CONSUMPTION_CURRENT > 0) & (df.ENERGY_CONSUMPTION_POTENTIAL > 0)]
```

```python
df_simple = pd.concat(
    [
        df['UPRN'],
        df['ENERGY_CONSUMPTION_CURRENT'] * df['TOTAL_FLOOR_AREA'],
        df['ENERGY_CONSUMPTION_POTENTIAL'] * df['TOTAL_FLOOR_AREA'],
        df['TOTAL_FLOOR_AREA']
    ],
    axis = 1,
    keys = ['UPRN', 'TotalEnergyConsumptionCurrent', 'TotalEnergyConsumptionPotential', 'TotalFloorArea']
)
df_simple
```

```python
df_simple['UPRN'] = df_simple['UPRN'].astype(int, errors = 'ignore')
df_sum = df_simple.groupby('UPRN').mean()
df_sum
```

```python
df_sum.to_csv('GroupedCertificates.csv', encoding='utf-8', index=True)
```
