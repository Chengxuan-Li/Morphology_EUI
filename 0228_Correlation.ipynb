```python
import momepy
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
buildings = gpd.read_file('files/attr/merged_buildings_attr.geojson')
```

```python
buildings.columns
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(buildings.dtypes)
```

```python
buildings.describe()
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(buildings.isnull().all())
```

```python
buildings = buildings.drop(columns=
                          ["ltbIBD", "sssLin"]
                          )
```

```python
buildings_dropna = buildings.dropna(inplace=False)
```

```python
buildings_dropna
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(buildings_dropna.dtypes)
```

```python
buildings_dropna = buildings_dropna.drop(columns=
                                        ["mapping_block_number", "age", "use", "data_level"]
                                        )
```

```python
buildings_dropna = buildings_dropna.drop(columns=
                                        ["USRN", "name"]
                                        )
```

```python
buildings_dropna = buildings_dropna.drop(columns=
                                        ["geometry"]
                                        )
```

```python
buildings_dropna = buildings_dropna.drop(columns=
                                        ["unique_property_number", 'unique_building_number', 'uID', 'bID', 'nID', 'eID','nodeID' ]
                                        )
```

```python
buildings_dropna.columns
```

```python
buildings_dropna = buildings_dropna.drop(columns=
                                        ["id", 'uID_int']
                                        )
```

```python
buildings_dropna["EUI"] = buildings_dropna["TotalEne_1_sum"]/buildings_dropna["TotalFloor_sum"]
```

```python
buildings_dropna = buildings_dropna.drop(columns=
                                        ['TotalEnerg_sum', 'TotalEne_1_sum']
                                        )
```

```python
buildings_dropna.columns
```

```python
buildings_dropna.to_csv("files/attr/raw_dataset.csv")
```

```python
normalized_df=(buildings_dropna-buildings_dropna.mean())/buildings_dropna.std()
```
